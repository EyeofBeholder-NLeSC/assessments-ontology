```python
from rdflib import Graph, URIRef, Literal
from rdflib.namespace import PROV, RDFS, SDO, RDF, SKOS, XSD
from rdflib.namespace import Namespace

from rdflib.tools.rdf2dot import rdf2dot

from IPython.display import display, Image
import pydotplus
import io

g = Graph()

DQV = Namespace("http://www.w3.org/ns/dqv#")
g.bind("dqv", DQV)

# define subjects
worker = URIRef(":worker")
iq_assessment = URIRef(":iq_assessment")
target_item = URIRef(":target_item")
completeness = URIRef(":completeness")
completenessMetric = URIRef(":completenessMetric")
# add triples starting from worker
g.add((worker, RDFS.subClassOf, SDO.Person))

# add triples starting from assessment
g.add((iq_assessment, RDFS.subClassOf, DQV.QualityMeasurement))
g.add((iq_assessment, PROV.wasAttributedTo, worker))
g.add((iq_assessment, DQV.computedOn, target_item))
g.add((iq_assessment, DQV.isMeasurementOf, completenessMetric))
g.add((iq_assessment, DQV.value, Literal('01', datatype=XSD.integer) ))

# add triples pointing to assessment
g.add((completeness, RDFS.subClassOf, DQV.dimension))
g.add((completeness, SKOS.prefLabel, Literal("Completeness","en")))
g.add((completeness, SKOS.definition, Literal("Completeness of the information","en")))
# add triples starting from target item
g.add((target_item, RDFS.subClassOf, SDO.DigitalDocument))

g.add((completenessMetric, RDFS.subClassOf, DQV.Metric))
g.add((completenessMetric, SKOS.definition, Literal("Likert scale value (ranging between -2 and +2) representing the completeness of the information item","en")))
g.add((completenessMetric, DQV.expectedDataType, XSD.integer))
g.add((completenessMetric, DQV.inDimension, completeness))


print(g.serialize(format='json-ld', indent=4))
```

```python
stream = io.StringIO()
rdf2dot(g, stream, opts={display})
dg = pydotplus.graph_from_dot_data(stream.getvalue())
png = dg.create_png()
display(Image(png))
```

```python
# quering with SPARQL
q = """
    SELECT ?a
    WHERE {
        ?a rdf:type dqv:dimension .
    }
"""

q_result = g.query(q)
for r in q_result:
    print("Label: ", g.value(r[0], RDFS.label))
    print("DType: ", g.value(r[0], RDFS.Datatype))
```
